```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Carregar dados
df = pd.read_csv('/home/rafelparracoelho/Área de Trabalho/VScode.py/analysis-turnover-rh/data/raw/WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
# Entender o dataset
print("Shape:", df.shape)
print("\nColunas:", df.columns.tolist())
print("\nTipos:", df.dtypes)
print("\nInfo básica:")
print(df.info())
```

```python
# Primeiras linhas
print("\nPrimeiras linhas:")
df.head()

# Valores faltantes
print("\nValores nulos:")
print(df.isnull().sum())

# Estatísticas descritivas
print("\nEstatísticas:")
df.describe()
```

```python
# Distribuição da variável target (Attrition)
print("\nDistribuição Attrition:")
print(df['Attrition'].value_counts())
print(df['Attrition'].value_counts(normalize=True))
```

```python
plt.figure(figsize=(8, 5))
df['Attrition'].value_counts().sort_values().plot(kind='barh')
plt.title('Distribuição de Rotatividade de Funcionários')
plt.xlabel('Número de Colaboradores')
plt.ylabel('Status de Rotatividade')
plt.tight_layout()
plt.show()

```
