```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Carregar dados
df = pd.read_csv('/home/rafelparracoelho/Área de Trabalho/VScode.py/analysis-turnover-rh/data/raw/WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
# Distribuição da variável target (Attrition)
print("\nDistribuição Attrition:")
print(df['Attrition'].value_counts())
print(df['Attrition'].value_counts(normalize=True))
```

```python
# Contagem original
counts = df['Attrition'].value_counts()

# Mapeamento para PT-BR
labels_pt = counts.index.map({'Yes': 'Sim', 'No': 'Não'})

# Cores (Não em cinza, Sim destacado)
colors = [
    '#B0B0B0' if label == 'No' else '#2E1A87'
    for label in counts.index
]

plt.figure(figsize=(9, 5))

ax = counts.plot(
    kind='bar',
    color=colors,
    edgecolor='black',
    width=0.55
)

# Títulos e labels
plt.title('Distribuição de Rotatividade de Funcionários', fontsize=14, weight='bold')
plt.xlabel('Status de Rotatividade', fontsize=11)
plt.ylabel('Número de Colaboradores', fontsize=11)

# 🔹 Aplicar labels traduzidos
ax.set_xticklabels(labels_pt, rotation=0)

# 🔹 Espaço no eixo Y
max_value = counts.max()
ax.set_ylim(0, max_value * 1.15)

# 🔹 Valores acima das barras
for i, v in enumerate(counts):
    ax.text(i, v + max_value * 0.03, str(v), ha='center', fontsize=10)

# Design clean
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
